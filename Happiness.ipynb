Đọc dữ liệu

```python
import pandas as pd
df = pd.read_csv('/content/ggdrive/MyDrive/PROJECT MD3/2019.csv')
df
```

Làm sạch dữ liệu

```python
df.drop_duplicates(inplace = True)
df
```

File dữ liệu csv này không có hàng trùng lắm cùng các ô dữ liệu trống vì vậy File csv này không cần làm sạch

Các số liệu thống kê đặc tả cho dữ liệu

```python
df.drop(columns =['Overall rank']).describe()
```

Top10 nơi hạnh phúc nhất

```python
top_10 = df.nlargest(10,'Score')
print("Top 10 nơi hạnh phúc nhất thế giới: ")
top_10
```

Top 10 nơi kém hạnh phúc nhất thế giới

```python
bottom_10 = df.nsmallest(10,'Score')
print("Top 10 quốc gia kém hạnh phúc nhất thế giới: ")
bottom_10
```

Chênh lệch về GDP per capita, Social support, và các yếu tố khác giữa hai nhóm này.

```python
columns_to_compare = ['Score','GDP per capita', 'Social support',
                      'Healthy life expectancy', 'Freedom to make life choices',
                      'Generosity', 'Perceptions of corruption']
top_10_avg = top_10[columns_to_compare].mean()
bottom_10_avg = bottom_10[columns_to_compare].mean()
difference = top_10_avg - bottom_10_avg
difference = difference[columns_to_compare]
print("Chênh lệch trung bình giữa Top 10 và Bottom 10 quốc gia:")
print(difference)
```

Biểu đồ trực quan hóa

```python
import matplotlib.pyplot as plt
difference.sort_values(ascending = True)
plt.figure(figsize=(6,4))
difference.plot(kind='bar', color='skyblue')

plt.title("Chênh lệch trung bình giữa Top 10 và Bottom 10 quốc gia", fontsize=10)
plt.ylabel("Chênh lệch trung bình", fontsize=10)
plt.xticks(rotation=45, fontsize = 8)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Xác định yếu tố ảnh đóp góp lớn nhất đến score về chỉ số hạnh phúc

```python
columns_to_compare = ['Score','GDP per capita', 'Social support',
                      'Healthy life expectancy', 'Freedom to make life choices',
                      'Generosity', 'Perceptions of corruption']
correlation = df[columns_to_compare].corr()["Score"].sort_values(ascending=False)
print("Hệ số tương quan giữa Score và các chỉ số:")
print(correlation)
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
correlation_matrix = df[['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy',
                         'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Tương quan giữa các biến", fontsize=12)
plt.show()
```

nhóm các quốc gia theo GDP per capita và phân tích sự khác biệt trong Social support

```python
df['GDP_group'] = pd.qcut(df['GDP per capita'], q=3, labels=['Low', 'Medium', 'High'])
group_means = df.groupby('GDP_group', observed = False)['Social support'].mean()

# Vẽ biểu đồ cột so sánh
group_means.plot(kind='bar', color=['skyblue', 'orange', 'green'], figsize=(6,4))

plt.title('So sánh Social support giữa các nhóm GDP', fontsize=14)
plt.xlabel('Nhóm GDP', fontsize=12)
plt.ylabel('Social support (Trung bình)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

nhóm các quốc gia theo GDP per capita và phân tích sự khác biệt trong Healthy life expectancy

```python
df['GDP_group'] = pd.qcut(df['GDP per capita'], q=3, labels=['Low', 'Medium', 'High'])
group_means = df.groupby('GDP_group', observed = False)['Healthy life expectancy'].mean()

# Vẽ biểu đồ cột so sánh
group_means.plot(kind='bar', color=['skyblue', 'orange', 'green'], figsize=(6,4))

plt.title('So sánh Healthy life expectancy giữa các nhóm GDP', fontsize=14)
plt.xlabel('Nhóm GDP', fontsize=12)
plt.ylabel('Healthy life expectancy (Trung bình)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Mối quan hệ tương quan giữa score và perceptions of corruption

```python
plt.figure(figsize=(6, 4))

# Vẽ biểu đồ scatter với đường hồi quy
sns.regplot(
    data=df,
    x='Perceptions of corruption',
    y='Score',
    scatter_kws={'color': 'blue', 'alpha': 0.6},
    line_kws={'color': 'red'},
    ci=95
)

# Thiết lập tiêu đề và nhãn
plt.title("Mối tương quan giữa Score và Perceptions of Corruption", fontsize=12)
plt.xlabel("Perceptions of Corruption", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(alpha=0.3)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
